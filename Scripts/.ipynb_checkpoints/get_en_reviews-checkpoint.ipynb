```python
import pandas as pd

from langdetect import detect
```

```python
# path to your file
df = pd.read_csv('./samerica/buenosaires.csv')
```

```python
df2 = df[df.comments.str[:34] != "The host canceled this reservation"]
df2 = df2.copy()
df2
```

```python
df3 = df2['comments']
```

```python
langs = []
for i in df3:
    try:
        langs.append(detect(i))
    except:
        langs.append('Nan')
        continue
```

```python
df2['langs'] = langs
df_en = df2.loc[df2['langs'] == 'en'].copy()
df_en
```

```python
# change the name of the resulting file
df_en.to_csv(r'en_reviews/buenosaires_en.csv')
```
